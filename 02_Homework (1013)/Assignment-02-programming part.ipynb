```python
# 建立站点之间连接
def build_station_connection(line_station, lines):
    station_connection = defaultdict(list)
    
    for l in lines:
        for s in range(len(line_station[l])):
            if s == 0:
                station_connection[line_station[l][s]].append(line_station[l][s+1])
            elif s == len(line_station[l])-1:
                station_connection[line_station[l][s]].append(line_station[l][s-1])
            else:
                station_connection[line_station[l][s]].append(line_station[l][s+1])
                station_connection[line_station[l][s]].append(line_station[l][s-1])
    
    return station_connection
```

#### 3. Build the search agent

建立搜索函数

Build the search agent based ont he graph you built.

for example, if you use Beijing subway graoh, and you run:

\>>> search("奥体中心“，”天安门“）

You should get the result as follows: 奥体中心 -> A ->B ->C ... -> 天安门

```python
def bfs(subway_map, start, destination): # 地图，起始点，终止点
    pathes = [[start]]
    visited = set()
    
    while pathes:
        path = pathes.pop(0) #拿出第一个点
        froniter = path[-1] #尾巴点
        
        if froniter in visited:continue
            
        successors = subway_map[froniter]
        
        for station in successors:
            if station in path : continue # 检查环
                
            new_path = path + [station]
            
            pathes.append(new_path)
            
            if station == destination:
                return new_path
            
        visited.add(froniter)
```

```python
def dfs(subway_map, start, destination): # 图，起始点，终止点
    pathes = [[start]]
    visited = set()
    
    while pathes:
        path = pathes.pop(0) #拿出第一个点
        froniter = path[-1] #尾巴点
        
        if froniter in visited:continue
            
        successors = subway_map[froniter]
        
        for station in successors:
            if station in path : continue # 检查环
                
            new_path = path + [station]
            
            pathes = [new_path] + pathes
            
            if station == destination:
                return new_path
            
        visited.add(froniter)
```

```python
# 选择最优路径的 BFS
def bfs_opi (subway_map,trans, station_line,start, destination,search_strategy): 
    pathes = [[start]]
    
    while pathes:
        path = pathes.pop(0) #拿出第一个点
        froniter = path[-1] #尾巴点
            
        successors = subway_map[froniter]
        
        for station in successors:
            if station in path : continue # 检查环
                
            new_path = path + [station]
            
            pathes.append(new_path)
            
        pathes = search_strategy(pathes,trans, station_line) # 排序
        
        if pathes and pathes[0][-1] == destination:
            return pathes[0]
```
